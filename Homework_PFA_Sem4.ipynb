## Условие 1: Задача 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('kc_house_data.csv', sep=',', encoding ='cp1251', quoting=3)
df.head()
```

- id - Уникальный ID для каждого дома
- date - Дата продажи дома
- price - Стоимость продажи дома
- bedrooms - Кол-во спален
- bathrooms - Кол-во ванных комнат (0.5 - туалет без душа)
- sqft_living - Кв. метры жилые
- sqft_lot - Кв. метры общие
- floors - Кол-во этажей
- waterfront - Есть набержная или нет
- view - Значение от 0 до 4 насколько хороший вид
- condition - Значение от 1 до 5 насколько хорошее состояние
- grade - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна
- sqft_above - Кв. метры дома, которые находятся выше земли
- sqft_basement - Кв. метры дома, которые находятся ниже земли
- yr_built - Год постройки дома
- yr_renovated - Год ремонта дома
- zipcode - Индекс
- lat - Широтма
- long - Долгота
- sqft_living15 - Кв. метры жилой площади у 15 соседей
- sqft_lot15 - Кв. метры общей площади у 15 соседей

* 1.1 Изучите стоимости недвижимости

```python
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена(млн.)')
plt.ylabel('Кол-во');
```

Вывод: Основная часть домов была постоена после 1950 года и до 2015 года

## Условие 2: 2 задача

* 2.1 Изучите распределение домов от наличия вида на набережную
* Постройте график
* Сделайте выводы

```python
data = df['waterfront'].value_counts()
```

```python
plt.pie(data.values, labels=data.index, autopct='%.2f%%');
```

Вывод: Домов с видом на набережную менее 1%.

* 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
```

```python
plt.pie(data.values, labels=data.index, autopct='%.2f%%');
```

```python
data = df['floors'].value_counts()
plt.figure(figsize=(4,4))
plt.bar(data.index, data.values)
plt.xticks(rotation=50);
```

Вывод: Большую часть составляют дома с 1м и 2мя этажами. Также много домов с этажностью 1,5. Все прочие дома составляют малую долю.

* 2.2 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
```

```python
plt.pie(data.values, labels=data.index, autopct='%.2f%%');
```

```python
data = df['condition'].value_counts()
plt.figure(figsize=(4,4))
plt.bar(data.index, data.values)
plt.xticks(rotation=50);
```

Вывод: Как видно из графика, основная часть домов находится в среднем состоянии 64,92%, в хорошем 26,28%б в отличном 7,87%. В плохом состоянии находится малая часть домов.

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['view'], y=df['price'])
plt.title('Изучите взаимосвязь стоимости недвижимости и наличия хорошего вида')
plt.xlabel('Наличие хорошего вида')
plt.ylabel('Стоимость, в млн.');
```

Вывод: Наличие хорошего вида очень влияет на стоимость недвижимости.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['floors'], y=df['price'])
plt.title('Изучите взаимосвязь стоимости недвижимости и этажности')
plt.xlabel('Этажность')
plt.ylabel('Стоимость, в млн.');
```

Вывод: Дома с этажностью 2,5 в целом стоят дороже, чем остальные дома.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['grade'], y=df['price'])
plt.title('Изучите взаимосвязь стоимости недвижимости и оценки')
plt.xlabel('Оценка')
plt.ylabel('Стоимость, в млн.');
```

Вывод: Чем выше оценка, тем выше стоимость.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['condition'], y=df['price'])
plt.title('Изучите взаимосвязь стоимости недвижимости и состояния')
plt.xlabel('Оценка состояния')
plt.ylabel('Стоимость, в млн.');
```

Вывод:Чем выше оценка состояния, тем выше стоимость недвижимости.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['bedrooms'], y=df['price'])
plt.title('Изучите взаимосвязь стоимости недвижимости и количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость, в млн.');
```

Вывод: На стоимость недвижимости влияет количество спален. Чем спален больше, тем дороже недвижимость.
