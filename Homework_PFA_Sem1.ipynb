## Условие 2: 
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

```python
fr = '''
Бананы - 5
Яблоки - 34
Груши - 55
Сливы - 69
Орехи - 234
'''
 
import re
 
def sum_fruit(d):
    return sum([int(i.group()) for i in re.finditer(r'\d+', d)])
 
print('Всего: ', sum_fruit(fr))
```

## 4.2 
Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
date = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09',
        '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26',
        '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27',
        '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01',
        '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31',
        '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
revenue = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471,
           651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519,
           3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]
d = dict.fromkeys(date) | dict(zip(date, revenue))
print(d)
```

```python
print(sum([m for d, m in zip(date, revenue) if d[5:7] == "11"]))
```

```python
def sum_in_month(date: list, revenue: list) -> dict:
    res = {}
    for d, m in zip(date, revenue):
        if d[5:7] in res.keys():
            res[d[5:7]] = res[d[5:7]] + m
        else:
            res[d[5:7]] = m
    return res
 
print(sum_in_month(date, revenue))
```
